```python
import pandas as pd 
import numpy as np
import sys
import os
```

```python
agree = pd.read_csv('data/datasets/D_target.csv')
```

```python
clients = pd.read_csv('data/datasets/D_clients.csv')
```

```python
clients
```

```python
total_df = agree.merge(clients, on='ID', how='left')
```

```python
salary = pd.read_csv('data/datasets/D_salary.csv')
```

```python
salary.columns = ['FAMILY_INCOME', 'PERSONAL_INCOME', 'ID']
```

```python
loans = pd.read_csv('data/datasets/D_loan.csv')
close_loans = pd.read_csv('data/datasets/D_close_loan.csv')
```

```python
loans = loans.merge(close_loans, on='ID_LOAN', how='left')
```

```python
loans_clients = loans.groupby('ID_CLIENT').count()
```

```python
loans.groupby('ID_CLIENT').sum()
```

```python
loans_clients['closed_loans'] = loans.groupby('ID_CLIENT').sum()['CLOSED_FL']
```

```python
loans_clients
```

```python
loans_clients['ID'] = loans_clients.index
```

```python
loans_clients = loans_clients.drop(columns=['CLOSED_FL'])
```

```python
loans_clients.columns = ['NUM_LOANS', 'NUM_CLOSED_LOANS', 'ID']
```

```python
total_df = total_df.merge(loans_clients, on='ID', how='left')
```

```python
total_df = total_df.drop_duplicates(subset=['AGREEMENT_RK'])
```

```python
total_df.isna().sum()
```

```python
total_df.to_csv('data/result.csv', index=0)
```
